Create a DFS to traverse the tree and change direction of edges if needed </br>

1. create adjacent list to represent the tree 
2. assign direction to each edge: i->j or j->"-i"  if"+": need to change direction
3. traverse from 0, if adjacent node not visited, do dfs again</br> 
    if +, need to change dirction (dfs is from root to leaf, here path requires leaf to root0)


Time complexity: O(n), where n is the number of nodes in the tree. We traverse the tree once using DFS. </br>

Space complexity: O(n), where n is the number of nodes in the tree. We use a boolean array to keep track of visited nodes. Also, we use an adjacency list to represent the tree, which requires O(n) space.

```python
class Solution:
    def dfs(self, adj, visited, startnode):
        num_change = 0
        visited[startnode] = True
        for adj_node in adj[startnode]:
            if not visited[abs(adj_node)]:#abs(adj_node)
                if adj_node > 0: 
                    #direction is from startnode to adj_node, need to change direction
                    num_change += 1
                num_change += self.dfs(adj,visited,abs(adj_node)) #abs(adj_node)
        return num_change


    def minReorder(self,n,connections):
        adj = [[] for _ in range(n)]
        for c in connections:
            adj[c[0]].append(c[1])
            adj[c[1]].append(-c[0])
        visited = [False for _ in range(n)]
        num_change = self.dfs(adj,visited,0)
        return num_change
```

This year, there will be a big event in the capital (city 0), and many people want to travel to this city.

It's guaranteed that each city can reach city 0 after reorder.
