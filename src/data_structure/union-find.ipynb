```python
class UF:
    # 记录连通分量
    def __init__(self, n: int):
        # 一开始互不连通
        self.count = n
        # 父节点指针初始指向自己
        self.parent = [i for i in range(n)]
        
        #平衡性优化
        #希望小一些的树接到大一些的树下面，避免头重脚轻\
        #解决方法是额外使用一个 size 数组，记录每棵树包含的节点数，我们不妨称为「重量」：
        self.size = [1]*n

    """将 p 和 q 连接"""
    def union(self, p: int, q: int) -> None:
        rootP = self.find(p)
        rootQ = self.find(q)
        if rootP == rootQ:
            return
        # 将两棵树合并为一棵
        # 小树接到大树下面，平衡优化
        if self.size[rootP] > self.size[rootQ]:
            self.parent[rootQ] = rootP
            self.size[rootP] += self.size[rootQ]
        else:
            self.parent[rootP] = rootQ
            self.size[rootQ] += self.size[rootP]
        self.count -= 1

    """ 返回某个节点 x 的根节点 """
    def find(self, x: int) -> int:
        # # 基础：根节点的 parent[x] == x
        # while self.parent[x] != x:
        #     x = self.parent[x]
        # return x

        #路径优化 
        # # plan1. 递归
        # if self.parent[x] !=x:
        #     self.parent[x] = self.find(self.parent[x])
        # return self.parent[x]
    
        # # plan2. 所有节点直接接到root下 (平衡优化不需要)
        # root = x
        # while self.parent[root]!=x:
        #     root = self.parent[x]
        # old_parent = self.parent[x]
        # # 然后把 x 到根节点之间的所有节点直接接到根节点下面
        # while x != root:
        #     self.parent[x] = root
        #     x = old_parent
        #     old_parent = self.parent[old_parent]
        # return root


        #plan3. 一次循环做两部，这个最快
        while self.parent[n] != n:
            self.parent[n] = self.parent[self.parent[n]]
            n = self.parent[n]
        return n #这种写法超快

    """判断 p 和 q 是否连通"""
    """返回图中有多少个连通分量"""
    def connected(self, p: int, q: int) -> bool:
        root_p = self.find(p)
        root_q = self.find(q)
        return root_p == root_q
```
