```python
# 1. BFS
class Solution_BFS:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        # 值传递-递归三要素
        # 1.状态语义： 状态点-课程，点数=numCourses
        #      边界状态：indegree=0的课程  目标状态：出队数量
        # 2.递推公式：无值传递，不需要dp
        # 3.传播方向：先序传播，已知-位置，waiting queue

        # BFS
        # 0. 存储数据，构建邻接图
        # 1. 统计入度
        graph = [[] for _ in range(numCourses)]
        indeg = [0]*numCourses
        for edge in prerequisites:
            graph[edge[1]].append(edge[0])
            indeg[edge[0]] += 1

        # 2. 构建QUEUE
        queue = []
        for i in range(numCourses):
            if indeg[i] == 0:
                queue.append(i)

        # *检查有无成环：统计出队数量
        outNum = 0
        while queue:
          # 3. 出队
            x = queue.pop(0)
            outNum += 1

            # 4. 扩展 # 5. 入队
            for y in graph[x]:
                indeg[y] -= 1
                if indeg[y] == 0:
                    queue.append(y)

        return outNum == numCourses
```

### DFS-postorder

1. 先序：path → 对 path 的简化 → 用回溯优化
2. 后序：memo → 等价于动规中的记忆化，相当于是当前节点的值已经被计算出来了！！！
3. 注意：path 检测要在 memo 检测之前做才行！！！


```python
# 2. DFS-postorder
class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        # 1. 建图： 邻接矩阵，有向图
        graph = [[] for _ in range(numCourses)]
        for edge in prerequisites:
            graph[edge[1]].append(edge[0])

        # 2. 深度搜索
        cycle = False
        path = [False]*numCourses  # 判圈剪枝：正在进行
        memo = [False]*numCourses  # 访问剪枝：已经结束 -> 有时访问剪枝需要携带访问完毕的返回结果
        topo = []  # 标记处理完成点

        def dfs(x, cycle, path, memo, topo):

            # 成环，直接结束
            if cycle is True:
                return 1

            # 在dfs(x)还没结束的时候又来了一个dfs(x),代表有圈
            if path[x] is True:
                print("test")
                cycle = True
                return 1

            # 判重：dfs(x)已经完成，不用再做一次
            if memo[x] is True:
                return 10  # 代表dfs(x)已经发生过一次，不用再来一次了

            path[x] = True  # 等价于path.push(x) → 回溯的思想

            for node in graph[x]:
                temp = dfs(node, cycle, path, memo, topo)  # 这里可以嵌入各种递推公式
                if temp == 1:
                    cycle = True
                    return 1

            path[x] = False  # 等价于path.pop()   → 回溯的思想

            memo[x] = True  # 缓存思想：代表从该节点出发的后序结点全部访问完毕，注意这里可以做记忆化

            topo.append(x)

        # 出发点：直接遍历所有点
        for i in range(numCourses):
            temp = dfs(i, cycle, path, memo, topo)
            if temp == 1:
                cycle = True

        if cycle:
            print("1")
            return False
        else:
            print("2")
            return len(topo) == numCourses
```

```python
# 2. DFS-postorder
class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:

        graph = [[] for _ in range(numCourses)]
        for edge in prerequisites:
            graph[edge[1]].append(edge[0])

        cycle = False
        path = [False]*numCourses
        memo = [False]*numCourses
        topo = []

        def dfs(x):
            nonlocal cycle
            nonlocal path
            nonlocal memo
            nonlocal topo

            if cycle is True:
                return

            if path[x] is True:
                cycle = True
                return

            if memo[x] is True:
                return memo[x]

            path[x] = True

            for node in graph[x]:
                dfs(node)

            path[x] = False

            memo[x] = True

            topo.append(x)

        for i in range(numCourses):
            dfs(i)

        if cycle:
            print("1")
            return False
        else:
            print("2")
            return len(topo) == numCourses
```
