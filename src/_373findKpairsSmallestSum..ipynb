```python
# priority queue
#  O((n1+k)log(n1))
import heapq

class Solution:
    # use different way to avoid redundancy
    # method1: add all eles in num1 first, then iterate through num2;
    #          in heapq [sum value, position of num2]
     # method2: visited set(); adding ele from num1/2 in same iteration
    #          in heapq (sum value, position of num1, position of num2) 
    
    #Runtime 1244 ms Beats 20.1% Memory 42.8 MB Beats 9.46%
    def kSmallestPairs_1(self, nums1: List[int], nums2: List[int], k: int) -> List[List[int]]:
        res = []
        # Priority queue to store pairs with smallest sums, sorted by the sum
        pq = [] 

        n2 = nums2[0]
        # Push the initial pairs into the priority queue
        for num1 in nums1:
            heapq.heappush(pq,[num1+n2,0])

        # Pop the k smallest pairs from the priority queue
        while k>0 and pq:
          minpair = heapq.heappop(pq)
          sumval, index2 = minpair[0], minpair[1]
          res.append([sumval-nums2[index2], nums2[index2]]) #获取num1的值：sum-num2

          k -= 1 # decerase k!!!!!!!!!!!

          # If there are more elements in nums2, push the next pair into the priority queue
          if index2+1 < len(nums2):
              heapq.heappush(pq, [sumval-nums2[index2]+nums2[index2+1],index2+1])

        return res

    #Runtime 1130 ms Beats 55% Memory 36.5 MB Beats 24.36%
    def kSmallestPairs_2(self, nums1: List[int], nums2: List[int], k: int) -> List[List[int]]:
        if not nums1 or not nums2 or not k:
            return []
        
        res = []
        pq = []
        visited = set()

        heapq.heappush(pq, (nums1[0]+nums2[0], 0, 0))
        visited.add((0,0))

        while k and pq:
            sumval, i, j = heapq.heappop(pq)
            res.append([nums1[i],nums2[j]])

            k-=1

            if i+1<len(nums1) and (i+1,j) not in visited:
              heapq.heappush(pq,(nums1[i+1]+nums2[j], i+1, j ))
              visited.add((i+1,j))

            if j+1<len(nums2) and (i,j+1) not in visited:
              heapq.heappush(pq,(nums1[i]+nums2[j+1], i, j+1 ))
              visited.add((i,j+1))

        return res
        
```
