```python
# BFS

# notice: 
# 1. probability will result in precision erros: differs 1e-5
# 2. LOG probability


class Solution:
    def maxProbability(self, n: int, edges: List[List[int]], succProb: List[float], start: int, end: int) -> float:
        
        adj = [[] for _ in range(n)]
        for i in edges:
            u,v = edge[i]
            p = succProb[i]
            adj[u].append((v,p))
            adj[v].append((u,p))
        
        # distance array
        # store the maximum probabilities of reaching each node from the starting node. 
        # Set the distance of the starting node to 1.0 (indicating certainty) 
        # Set the distances of all other nodes to 0.0 (indicating uncertainty).
        dist = [0,0]*n
        dist[start] = 1.0

        #queue for BFS
        que = deque([start])   #[node1, n2, n3]
        # que = deque()  que.append(start)  # same

        while que:
            curr = que.popleft()
            
            for node, prob in adj[curr]:
                new_prob = dist[curr]*prob
            
                if new_prob > dist[node]:
                    dist[node] = new_prob
                    que.append(node)
        
        return dist[end]

```
