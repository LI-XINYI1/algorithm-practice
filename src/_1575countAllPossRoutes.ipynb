Notice: 
1. we can REPEAT/looping，as long as we have fuel!!!!!
2. fuel must be sufficient for movement
3. avoid REDUNDANCY computation:  results of SUBproblems stored in a MEMOIZATION TABLE

```python
#1. DFS and memo - 递归和备忘录： 
# TC 38.34%; Memory 16.9 MB 97.74%
class Solution:
    
        
    MOD = 1000000007

    # DFS(current location, remaining fuel)
    def dfs(self, locations, city, finish, remianFuel, memo):
      
      # dont =0, because we need check if current city == target, then update total
      if remianFuel < 0:
         return 0
      
      # avoid redundant calculation
      if memo[city][remianFuel] is not None:
         return memo[city][remianFuel]
      
      # if the current city is target finish destination: set total = 1
      total = 1 if city == finish else 0

      for next_city in range(len(locations)):
        # continue when
        # 1. not current city
        # 2. the remianing fuel is sufficient
        spentFuel =  abs(locations[next_city]-locations[city])
        if next_city != city and remianFuel >= spentFuel:
           total = (total + self.dfs(locations, next_city, finish, remianFuel - spentFuel, memo)) % self.MOD
        
      memo[city][remianFuel] = total

      return total

    def countRoutes(self, locations: List[int], start: int, finish: int, fuel: int) -> int:
      n = len(locations)
      memo = [[None]*(fuel + 1) for _ in range(n)]
      return self.dfs(locations,start,finish,fuel,memo)
         
```

```python
#1. DP dp数组和迭代
# TC 11.28%; Memory 17.3 MB 88.72%
class Solution:
    
    # dp五要素
    #1. def of dp[currentCity][remainingFuel]: number of routes from current state to reaching finish destination
    #2. inductive formula
    #    when  not goto current city & remianing fuel is sufficient
    #    dp[curCity][f] = (dp[curCity][f] + dp[nextCity][f-requireFuel]) % MOD
    #3. initialization of dp[]: when start is finish, at least one route for any remaing fuels
    #4. sequence: fuel: from 0 to full, every city(sequence not important)

    def countRoutes(self, locations: List[int], start: int, finish: int, fuel: int) -> int:
      MOD = 1000000007
      n = len(locations)
      
      # create DynamicProgramming matrix
      dp = [[0]*(fuel + 1) for _ in range(n)]
      # initialize dp: when start is at final : has at least 1 success subroute 
      for f in range(fuel+1):
         dp[finish][f] = 1
      
      for f in range(fuel+1):
         for curCity in range(n):
            for nextCity in range(n):
               requireFuel = abs(locations[nextCity] - locations[curCity])
               if nextCity != curCity and f >= requireFuel:
                  dp[curCity][f] = (dp[curCity][f] + dp[nextCity][f-requireFuel]) % MOD

      return dp[start][fuel]
         
```

You are given an array of distinct positive integers locations where locations[i] represents the position of city i. You are also given integers start, finish and fuel representing the starting city, ending city, and the initial amount of fuel you have, respectively.

At each step, if you are at city i, you can pick any city j such that j != i and 0 <= j < locations.length and move to city j. Moving from city i to city j reduces the amount of fuel you have by |locations[i] - locations[j]|. Please notice that |x| denotes the absolute value of x.

Notice that fuel cannot become negative at any point in time, and that you are allowed to visit any city more than once (including start and finish).

Return the count of all possible routes from start to finish. Since the answer may be too large, return it modulo 109 + 7.

 

Example 1:

Input: locations = [2,3,6,8,4], start = 1, finish = 3, fuel = 5
Output: 4
Explanation: The following are all possible routes, each uses 5 units of fuel:
1 -> 3
1 -> 2 -> 3
1 -> 4 -> 3
1 -> 4 -> 2 -> 3
Example 2:

Input: locations = [4,3,1], start = 1, finish = 0, fuel = 6
Output: 5
Explanation: The following are all possible routes:
1 -> 0, used fuel = 1
1 -> 2 -> 0, used fuel = 5
1 -> 2 -> 1 -> 0, used fuel = 5
1 -> 0 -> 1 -> 0, used fuel = 3
1 -> 0 -> 1 -> 0 -> 1 -> 0, used fuel = 5
Example 3:

Input: locations = [5,2,1], start = 0, finish = 2, fuel = 3
Output: 0
Explanation: It is impossible to get from 0 to 2 using only 3 units of fuel since the shortest route needs 4 units of fuel.
